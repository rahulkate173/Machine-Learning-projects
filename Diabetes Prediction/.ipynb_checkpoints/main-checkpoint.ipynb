### Here in this dataframe their are lot of outlier so, what i will do :
#### 1. get the range of df
#### 2. replace the outlier bu colums median to get the original shape
#### 3. apply standardscalar on the df

```python
def outlier(df):
    df_outlier=pd.DataFrame() # created new outlier free dataframe
    for col in df[:-1]: # olny skipping last outcome column
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        iqr=q3-q1
        lower=q1-(1.5*iqr)
        upper=q3+(1.5*iqr)
        median=df[col].median()
        df_outlier[col+'_']=df[col].apply(lambda x: median if x<lower or x>upper else x)

    df_outlier['Outcome_']=df['Outcome']
    return df_outlier
```

```python
df_noOutlier=outlier(df)
```

```python
from sklearn.preprocessing import StandardScaler
def scaled_df(df):
    scalar = StandardScaler()
    df_scaledDummy = pd.DataFrame()

    for col in df.columns[:-1]:  # assuming last column is 'Outcome'
        x = df[[col]]  # keep as DataFrame for scaler
        scaled_col = scalar.fit_transform(x).ravel()  # convert to 1D
        df_scaledDummy[col + '_scaled'] = scaled_col
    
    df_scaledDummy['Outcome_scaled'] = df['Outcome_']
    return df_scaledDummy
```

```python
df_scaled=scaled_df(df_noOutlier)
```
